```python
import pandas as pd
import os
import numpy as np
```

```python
anime = pd.read_csv('anime-recommendation-data/anime.csv')
animelist = pd.read_csv('anime-recommendation-data/animelist.csv')
```

```python
anime.rename(columns = {'Anime-PlanetID':'anime_id'}, inplace=True)
```

```python
all_anime_name = pd.merge(animelist, anime[['anime_id', 'Name']], on = 'anime_id', how = 'left')
```

```python
all_anime = pd.merge(all_anime_name, anime[['anime_id', 'Tags']], on = 'anime_id', how = 'left')
```
